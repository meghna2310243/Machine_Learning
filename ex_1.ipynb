```python
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    text = text.lower()
    text = ''.join([char for char in text if char not in string.punctuation])
    words = text.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    cleaned_text = ' '.join(words)
    return cleaned_text

df['cleaned_text'] = df['text'].apply(preprocess_text)
df
```

```python
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt

heights = []
widths = []

# Skip first invalid row
for path in df['image_path'][1:101]:
    img_path = Path(path)  # path is now correct

    if not img_path.exists():
        print("Image not found:", img_path)
        continue

    try:
        img = Image.open(img_path).convert("L")
        widths.append(img.width)
        heights.append(img.height)
    except Exception as e:
        print("Failed to read:", img_path, e)
        continue

# Plot histograms
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].hist(heights, bins=10, color='skyblue', edgecolor='black')
axes[0].set_title("Image Heights")
axes[0].set_xlabel("Pixels")
axes[0].set_ylabel("Count")

axes[1].hist(widths, bins=10, color='salmon', edgecolor='black')
axes[1].set_title("Image Widths")
axes[1].set_xlabel("Pixels")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt

# Folder containing images (relative to notebook)
img_folder = Path("../Img")

# Create subplots for 5 images
fig, axes = plt.subplots(1, 5, figsize=(15, 4))

# Skip first invalid row and take first 5 images
for i, path in enumerate(df['image_path'][1:6]):
    # Construct full path using Path
    img_path = img_folder / Path(path).name  # just the filename

    # Check if file exists
    if not img_path.exists():
        print("Image not found:", img_path)
        continue

    try:
        # Open image and convert to grayscale
        img = Image.open(img_path).convert("L")

        # Convert to a flat array of pixel values
        pixel_values = list(img.getdata())

        # Plot histogram
        axes[i].hist(pixel_values, bins=50, color='gray', edgecolor='black')
        axes[i].set_title(f"Image {i+1}")
        axes[i].set_xlabel("Pixel Value")
        axes[i].set_ylabel("Count")
    except Exception as e:
        print("Failed to read:", img_path, e)
        continue

plt.tight_layout()
plt.show()
```

```python
from pathlib import Path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img_folder = Path(r"Img")

# Get unique class labels
classes = df['label'].unique()

fig, axes = plt.subplots(1, len(classes), figsize=(5*len(classes), 4))

for i, label in enumerate(classes):
    imgs = []
    # Skip first row (invalid) and take first 20 images
    for path in df[df['label'] == label]['image_path'][1:21]:
        img_file = Path(path).name  # get just the filename
        img_path = img_folder / img_file

        if not img_path.exists():
            print("Image not found:", img_path)
            continue

        try:
            img = Image.open(img_path).convert("L")  # convert to grayscale
            img_array = np.array(img, dtype=np.float32)  # convert to float for averaging
            imgs.append(img_array)
        except Exception as e:
            print("Failed to read:", img_path, e)
            continue

    if len(imgs) == 0:
        print(f"No valid images found for class {label}")
        continue

    mean_img = np.mean(imgs, axis=0)
    axes[i].imshow(mean_img, cmap='gray')
    axes[i].set_title(f"Mean Class {label}")
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```
